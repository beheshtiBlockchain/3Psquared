```python
def find_notrep(x0, y0, x1, y1, x2, y2):
  ret = [0,0]
  if x0 == x1:
    ret[0] = x2
  elif x0 == x2:
    ret[0] = x1
  elif x1 == x2:
    ret[0] = x0
  if y0 == y1:
    ret[1] = y2
  elif y0 == y2:
    ret[1] = y1
  elif y1 == y2:
    ret[1] = y0
  return ret
x0y0 = input("")
x0y0a = x0y0.split(" ")
x0, y0 = int(x0y0a[0]), int(x0y0a[1])
x1y1 = input("")
x1y1a = x1y1.split(" ")
x1, y1 = int(x1y1a[0]), int(x1y1a[1])
x2y2 = input("")
x2y2a = x2y2.split(" ")
x2, y2 = int(x2y2a[0]), int(x2y2a[1])

result = find_notrep(x0,y0,x1,y1,x2,y2)
print(str(result[0])+" "+str(result[1]))
```

```python
x = input("")
xa = x.split(" ")
m, n = int(xa[0]), int(xa[1])
"""
m is black
n is white
"""
def khalaf(m, n):
  b, w = m, n
  common = min(b,w)
  b = b - common
  w = w - common
  b = b + common
  return b,w
def ham(m, n):
  b, w = m, n
  if(b == max(b, w)):
    w = w + int(b/2)
    b = b%2
  elif(w == max(b, w)):
    w = w % 2 + int(w/2)
    b = b
  return b,w
m0, n0 = m, n
while max(m,n)>1 and not m is n:
  print(m,n)
  if m%2 == 0 or n%2 == 0:
    m,n = ham(m,n)
  else:
    m, n = khalaf(m,n)
while max(m0,n0)>1 and not m0 is n0:
  if m0%2 != 0 and n0%2 != 0:
    m0, n0 = khalaf(m0,n0)
  else:
    m0,n0 = ham(m0,n0)
"""
m n m0 n0 F
0 0 0  0  np
0 0 0  1  np -conflict
0 0 1  0  np
0 0 1  1  elm
0 1 0  0  np
0 1 0  1  white - consensus
0 1 1  0  np
0 1 1  1  elm
1 0 0  0  np
1 0 0  1  np
1 0 1  0  black
1 0 1  1  elm
1 1 0  0  elm
1 1 0  1  elm
1 1 1  0  elm
1 1 1  1  elm
elm = eluminated
"""
print(str(m0)+str(m)+str(n0)+str(n))
if m0 and m and (not n0) and (not n):
  print("black")
elif n0 and n and (not m0) and (not m):
  print("white")
else:
  print("noprediction")
```

```python
import time
import numpy as np
n, d = map(int, input().split())
a1 = np.array(list(range(100))*int(n/100))
tic = time.time()
num = 0
maxs = 1000000000

for i in range(d):
  z = 0
  f = 0
  for j in range(i,n,d):
    if a1[j]%2 == 0:
      z+=1
    else:
      f+=1
  num+=min(z,f)
  if max(z,f)<maxs:
    maxs = max(z,f)
    m = min(z,f)
  if min(z,f) == z:
    for k in range(i,n,d):
      if a1[k]%2==0:
        a1[k]+=1
  else:
    for k in range(i,n,d):
      if not a1[k]%2==0:
        a1[k]+=1
iseven=True
for i in range(d):
  if not a1[i]%2==0:
    iseven = not iseven
if iseven:
  print(num)
else:
  print(num-m+maxs)
toc=time.time()
elapsed = toc-tic
print(elapsed)
```
